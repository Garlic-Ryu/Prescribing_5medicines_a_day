```python
# 출력 설정
def format_prescription_output_ordered(prescription_details, iteration):
    output = f"{iteration}회차\n"
    for category in ['간계', '심계', '비계', '폐계', '신계']:
        pres_name, symptoms = prescription_details[category]
        # 한글증상명을 문자열 형식으로 변환
        symptoms_str = ', '.join([f"['{symptom}', '{explanation}']" for symptom, explanation in symptoms])
        output += f"'{pres_name}', [{symptoms_str}]\n"
    return output
```
